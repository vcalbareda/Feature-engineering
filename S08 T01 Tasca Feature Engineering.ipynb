```python
raw_df = pd.read_csv('lol2021.csv')
raw_df
```

Se normaliza los datos de muertes y torretas contruidas

```python
ss = StandardScaler()

Xstd = ss.fit_transform(raw_df[['Kills', 'Wards Placed']].values)
Xstd
```

```python
print (Xstd.mean())
print (Xstd.std())
```
